```python
import pickle
from pymongo import MongoClient
from decouple import Config, config
import random
```

```python
f = open('books-data.pkl', 'rb')
data = pickle.load(f)
data
```

```python
book_shelves = 10
shelf_sections = ['A', 'B', 'C', 'D', 'E']
def add_book_location(book):
    book['location'] = {
        'shelf': random.randint(1,book_shelves),
        'section': random.choice(shelf_sections)
    }
    del book['Unnamed: 0']
    return book
```

```python
book_list = list(map(add_book_location,data))
```
